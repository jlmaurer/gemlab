```python
import matplotlib.pyplot as plt
import numpy as np
```

Anti-plane strain is often used to model interseismic strike-slip fault motion. The model is not too bad as long as you are in a fairly long and straight section of a strike-slip fault. By modeling the velocity and/or strain rate, we can determine the parameters that relate to fault locking (slip rate and locking depth).

```python
def antiplane_velocity(sdot, d, x):
    """Calculate and return predicted velocity for an antiplane strike-slip fault"""
    return (sdot / np.pi) * np.arctan2(x, d)
```

```python
def antiplane_strain(sdot, d, x):
    """
    Calculate anti-plane strain for an antiplane strike-slip fault

    Arguments:
      params:   list of slip rate (mm/yr) and locking depth (km)
      x:        Observation points

    Returns:
      strain rate in nanostrain/year
    """
    sdot = sdot / 1000  # convert to m/yr
    d = d * 1000  # convert to m
    x = x * 1000  # convert to m
    return 1e9 * (sdot / (2 * np.pi * d)) / (1 + np.square(x / d))
```

```python
x = np.linspace(-50, 50)
```

```python
params = {
    'sdot': 10,
    'd': 10,
}
```

```python
plt.plot(x, antiplane_velocity(**params, x=x), label='Velocity')
plt.xlabel('Distance (km)')
plt.ylabel('Velocity (mm/yr)')
```

```python
plt.plot(x, antiplane_strain(**params, x=x), label='Strain rate')
```
